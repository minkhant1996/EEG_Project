```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import mne
from mne import io
from mne import read_epochs
```

```python
url = './data/'

print("\n========= epoch_1 ==========")

# epoch_1
file_name_1 = 'g1dual1_epo.fif' # eg. 'g1dual1_epo.fif'
epochs = read_epochs(url + file_name_1)
```

```python
event_id = {'Non-Target': 1, 'Target': 2}
tmin = -0.1
tmax = 0.7
picks = np.arange(0, 32)
print(picks)
```

```python
evoked_non = epochs['Non-Target'].average()
evoked_non.plot(spatial_colors = True, gfp = True, time_unit = 's')
evoked_target = epochs['Target'].average()
evoked_target.plot(spatial_colors = True, gfp = True, time_unit = 's')
```

```python
epochs['Non-Target'].plot_image(picks = 'Cz')
epochs['Target'].plot_image(picks = 'Cz')
```

```python
epochs['Non-Target'].plot_image(picks = 'Pz')
epochs['Target'].plot_image(picks = 'Pz')
```

```python
total_evoked = {"non": evoked_non, "target": evoked_target}
mne.viz.plot_compare_evokeds(total_evoked, picks = picks, combine = 'mean')
```
